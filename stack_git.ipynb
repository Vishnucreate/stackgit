```python
import requests
import openai
import json

# Replace these with your API keys
GOOGLE_API_KEY = "your_api_key"
CX = "your_api_key"
OPENAI_API_KEY = "your_api_key"



def search_google(query):
    url = "https://www.googleapis.com/customsearch/v1"
    params = {
        "q": query,
        "key": GOOGLE_API_KEY,
        "cx": CX,
    }
    response = requests.get(url, params=params)
    if response.status_code == 200:
        return response.json()
    else:
        print("Error with Google Search API:", response.text)
        return None

# Step 2: Extract Relevant Data

def extract_search_results(data):
    """
    Extracts titles, snippets, and links from Google Search API results.
    """
    results = []
    if "items" in data:
        for item in data["items"]:
            title = item.get("title", "No Title")
            snippet = item.get("snippet", "No Snippet")
            link = item.get("link", "No Link")
            results.append({"title": title, "snippet": snippet, "link": link})
    return results

# Step 3: Summarize Results Using OpenAI

def summarize_results_with_openai(parsed_results, query):
    """
    Generate a summary from parsed results using OpenAI GPT.
    """
    if not parsed_results:
        return "No relevant results to summarize."

    content = "\n\n".join(
        f"{res['title']} - {res['snippet']} ({res['link']})" for res in parsed_results
    )
    prompt = (
        f"Summarize the following search results for the query: '{query}'\n\n"
        f"{content}\n\nSummary:"
    )

    try:
        response = openai.ChatCompletion.create(
            model="gpt-3.5-turbo",
            messages=[
                {"role": "system", "content": "You are a helpful assistant that summarizes search results."},
                {"role": "user", "content": prompt}
            ],
            max_tokens=300
        )
        return response['choices'][0]['message']['content'].strip()
    except Exception as e:
        return f"Error with OpenAI API: {e}"

# Main Function

def main():
    query = input("Enter your Python error message or query: ")
    print(f"Searching Google for: {query}...")

    # Step 1: Search Google
    data = search_google(query)

    if not data:
        print("No data received from Google Search API.")
        return

    # Step 2: Extract Results
    parsed_results = extract_search_results(data)

    if not parsed_results:
        print("No relevant results found.")
        return

    print("\nExtracted Results:")
    for idx, result in enumerate(parsed_results, 1):
        print(f"Result {idx}:")
        print(f"Title: {result['title']}\nSnippet: {result['snippet']}\nLink: {result['link']}\n")

    # Step 3: Summarize Results
    print("Summarizing results using OpenAI...")
    summary = summarize_results_with_openai(parsed_results, query)

    print("\nSummary of Results:")
    print(summary)

if __name__ == "__main__":
    main()

```
